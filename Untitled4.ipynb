```python


def calculate_distance_matrix(csv_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv("C:\matplotlib\dataset-3.csv", delimiter='\t')

    # Create a set of all unique IDs
    unique_ids = set(df['id_start'].unique()) | set(df['id_end'].unique())

    # Create an empty DataFrame for the distance matrix
    distance_matrix = pd.DataFrame(index=unique_ids, columns=unique_ids)

    # Fill diagonal values with 0
    distance_matrix.values[[range(len(unique_ids))]*2] = 0

    # Iterate over rows in the original DataFrame to fill in distance_matrix
    for index, row in df.iterrows():
        start, end, distance = row['id_start'], row['id_end'], row['distance']

        # Update the matrix with bidirectional distances
        distance_matrix.at[start, end] = distance_matrix.at[end, start] = distance

    # Calculate cumulative distances
    for i in unique_ids:
        for j in unique_ids:
            for k in unique_ids:
                if pd.notna(distance_matrix.at[i, k]) and pd.notna(distance_matrix.at[k, j]):
                    # Update the cumulative distance
                    distance_matrix.at[i, j] = distance_matrix.at[j, i] = (
                        distance_matrix.at[i, k] + distance_matrix.at[k, j]
                    )

    return distance_matrix

# Example usage
csv_file_path = "C:\matplotlib\dataset-3.csv"
resulting_matrix = calculate_distance_matrix("C:\matplotlib\dataset-3.csv")

# Print the resulting matrix
print(resulting_matrix)
```

```python


def unroll_distance_matrix(distance_matrix):
    # Create an empty DataFrame for unrolled distances
    unrolled_df = pd.DataFrame(columns=['id_start', 'id_end', 'distance'])

    # Iterate over rows in the distance matrix
    for i in distance_matrix.index:
        for j in distance_matrix.columns:
            if i != j:
                unrolled_df = unrolled_df.append({
                    'id_start': i,
                    'id_end': j,
                    'distance': distance_matrix.at[i, j]
                }, ignore_index=True)

    return unrolled_df

# Example usage with the resulting matrix from Question 1
resulting_matrix_from_q1 = calculate_distance_matrix("C:\matplotlib\dataset-3.csv")
unrolled_distances = unroll_distance_matrix(resulting_matrix_from_q1)

# Print the unrolled distances DataFrame
print(unrolled_distances)
```

```python
import pandas as pd

def find_ids_within_ten_percentage_threshold(df, reference_value):
    # Filter DataFrame to include only rows with the specified reference_value as id_start
    reference_df = df[df['id_start'] == reference_value]

    # Calculate the average distance for the reference value
    average_distance = reference_df['distance'].mean()

    # Calculate the lower and upper bounds within 10% of the average distance
    lower_bound = average_distance - (average_distance * 0.1)
    upper_bound = average_distance + (average_distance * 0.1)

    # Filter DataFrame to include only rows within the 10% threshold
    within_threshold_df = df[(df['distance'] >= lower_bound) & (df['distance'] <= upper_bound)]

    # Extract and sort unique values from the id_start column
    result_ids = sorted(within_threshold_df['id_start'].unique())

    return result_ids

# Example usage with the unrolled distances DataFrame from Question 2
resulting_unrolled_distances = unroll_distance_matrix(resulting_matrix_from_q1)
reference_value = 1001402  # Replace with the desired reference value

# Find IDs within 10% threshold
result_ids = find_ids_within_ten_percentage_threshold(resulting_unrolled_distances, reference_value)

# Print the result
print("IDs within 10% threshold:", result_ids)
```

```python


def calculate_toll_rate(df):
    # Create columns for different vehicle types with their rate coefficients
    df['moto'] = 0.8
    df['car'] = 1.2
    df['rv'] = 1.5
    df['bus'] = 2.2
    df['truck'] = 3.6

    # Calculate toll rates by multiplying the distance with the rate coefficients for each vehicle type
    df['moto_toll'] = df['distance'] * df['moto']
    df['car_toll'] = df['distance'] * df['car']
    df['rv_toll'] = df['distance'] * df['rv']
    df['bus_toll'] = df['distance'] * df['bus']
    df['truck_toll'] = df['distance'] * df['truck']

    return df

# Example usage with the unrolled distances DataFrame from Question 2
resulting_unrolled_distances = unroll_distance_matrix(resulting_matrix_from_q1)

# Calculate toll rates
result_with_toll_rates = calculate_toll_rate(resulting_unrolled_distances)

# Print the resulting DataFrame with toll rates
print(result_with_toll_rates)
```
